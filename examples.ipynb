```python
# Example 1: Name Concatenation
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
full_name = first_name + " " + last_name
print(f"Full name: {full_name}")
```

```python
# Example 2: Area of a rectangle
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
area = length * width
print(f"Area of the rectangle: {area}")
```

```python
# Example 3: Even or Odd
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
# Example 4: Age comparison
age1 = int(input("Enter the age of the first person: "))
age2 = int(input("Enter the age of the second person: "))

if age1 > age2:
    print("The first person is older.")
elif age1 < age2:
    print("The second person is older.")
else:
    print("Both persons are the same age.")
```

```python
# Example 5: Blastoff
start = 10

while start > 0:
    print(start)
    start -= 1

print("Blast off!")
```

```python
# Example 6: Multiplication table

num = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
# Example 7: List Manipulation
numbers = [4, 8, 15, 16, 23, 42]
sum_numbers = sum(numbers)
average = sum_numbers / len(numbers)

print("Sum:", sum_numbers)
print("Average:", average)
```

```python
# Example 8: List Reversal
numbers = [1, 2, 3, 4, 5, 6]
reversed_numbers = []

for num in reversed(numbers):
    reversed_numbers.append(num)

print("Reversed list:", reversed_numbers)
```

```python
# Example 9: Our first plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y);
```

```python
# Example 10: Monthly Expenses Pie Chart
import matplotlib.pyplot as plt

categories = ['Rent', 'Food', 'Transportation', 'Entertainment', 'Savings']
expenses = [1200, 600, 150, 200, 450]

plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Monthly Expenses Distribution')
plt.show()
```
